```python

#2주전의 파일을 가져와서 복사 
import shutil
from datetime import date, datetime
import os
import re

#올해는 몇년도인가요?#########################################################(직접 작성)
now_y="25년"

#연도가 바뀌면 1월 첫째주와 둘째주 데이터는 가져올 수 없으므로 수작업 필요
from datetime import datetime, timedelta
origin_str = (datetime.now() - timedelta(weeks=2) - timedelta(days=1)).strftime("%m.%d")
#임시로 고정
origin_str = '2차서식'
#print("2주전 날짜:",origin_str)  # 예: 2025-07-26

#임시로 고정 (작업해야하는 파일의 날짜) ###########################################################작업 시 수정
now_str = input("작업할 날짜를 넣어주세요")#"09.02"
print("작업할 파일의 날짜:",now_str)

# 이번달 
now_m = str(int(now_str[:2]))+"월"

source_folder = r".\{0}\{1}\{2}".format(now_y,now_m,origin_str) #메인경로 수정
print("가져올 폴더:",source_folder)
```

```python
#폴더 생성 전에 해당 월의 폴더가 있는지 확인 
if os.path.exists(r"./{0}\{1}".format(now_y, now_m)):
    print("폴더가 존재합니다.")
else:
    print("폴더가 없습니다.")
    os.mkdir(r"./{0}\{1}".format(now_y, now_m))

# 복사할 대상 경로
destination_folder = r".\{0}\{1}\{2}.{3}".format(now_y, now_m, str(int(now_str[:2])),str(int(now_str[3:]))) #메인경로 수정 
print("복사한 데이터를 붙일 폴더:",destination_folder)

#오늘날짜 폴더 생성
# 폴더 생성 (이미 있으면 에러 발생)
# 폴더 복사
shutil.copytree(source_folder, destination_folder)
print("폴더 복사가 완료되었습니다!")
```

| 항목             | `Dispatch`                 | `EnsureDispatch`            |
| -------------- | -------------------------- | --------------------------- |
| 타입 라이브러리 캐시 생성 | ❌ 안 함                      | ✅ 생성                        |
| 속성/메서드 자동완성    | 거의 불가능 (런타임 호출)            | 일부 가능 (타입 정보 로드됨)           |
| 초기 속도          | 더 빠름 (바로 호출)               | 처음은 느림 (gencache 생성)        |
| 안정성            | COM 서버가 준비 안 됐을 때 예외 발생 가능 | gencache를 통해 COM 타입 로딩 후 연결 |
| 쓰임새            | 단순 실행, 속도 우선               | 개발·디버깅, 코드 자동완성, 안정성 우선     |
